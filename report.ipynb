```python
## preprocessing 
import pandas as pd
from textblob import TextBlob
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression ,LinearRegression

from sklearn.feature_extraction.text import TfidfVectorizer
data1 = pd.read_csv("chelseafc.csv").fillna(" ")
x =data = data1["comments"]
y = data1["likes"]
```

```python
word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='word',
    ngram_range=(1, 5),
    max_features=1000)

word_vectorizer.fit(data)
```

## Logistic regression

```python
# len(char_vectorizer.get_feature_names())
# char_vectorizer.inverse_transform(train_char_features)
trainx , testx ,trainy ,testy = train_test_split(x , y ,test_size=0.2)
train_features = word_vectorizer.transform(trainx)
test_features = word_vectorizer.transform(testx)
print(word_vectorizer.inverse_transform(train_features))
model = LogisticRegression(solver='sag',multi_class="ovr",tol=1e-10)
model.fit(train_features,trainy)
predicted_likes = model.predict(test_features)
error = [ abs(i-j) for (i,j) in zip(predicted_likes,testy)]
```

```python
print("average fluctutaion  in likes",sum(error)/len(testy))
```

# Linear regression
